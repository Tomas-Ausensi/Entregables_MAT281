##Districts

Recordemos el dataset

```python
districts.head(1)
```

- **district_id**: El identificador único del distrito escolar
- **State**: El estado donde se ubica el distrito
- **locale**: Clasificación de la NCES que categoriza el territorio de los Estados Unidos en cuatro tipos: Ciudad, Suburbio, Pueblo y Rural.
- **pct_black/hispanic**: Porcentaje de estudiantes en los distritos identificados como Negros o Hispanicos. Basado en los datos de la NCES durante 2018-2019
- **pct_free/reduced** Porcentaje de estudiantes en los distritos que son candidatos para una beca alimenticia (especifícamente, para el almuerzo) parcial o total. Basado en los datos de la NCES durante 2018-2019
- **County_connections_ratio**: ratio (Conexiones de alta velocidad residencial fija sobre los 200 kilobits por segundo en al menos una dirección/hogar).
- **pp_total_raw**: gasto por pupilo (suma del gasto local y federal) de Edunomics Lab's National Education Resource Database on Schools project. La información del gasto es por cada escuela. Se usa el valor medio para representar el gasto para un distrito dado

```python
index = districts.state.value_counts().index
ax = sns.countplot(data= districts, y = 'state', order = index, )

total = len(districts['state'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width()/total)
    x = p.get_x() + p.get_width() + 0.1
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y), ha='left', va='center')

""" for vertical countplots
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2 - 0.1
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')
"""
plt.title('Estados')
plt.show()
```

Como se puede observar, la mayor representación la tiene el estado de Connecticut, seguido de Utah y Massachusets

```python
#Extracted from https://www.kaggle.com/code/assurivyshnavi/digital-learning
fig, ax  = plt.subplots(figsize=(8, 5))
fig.suptitle('Locale Type Distribution', size = 5)

labels = list(districts.locale.value_counts().index)
sizes = districts.locale.value_counts().values

ax.pie(sizes, startangle=60, labels=labels,autopct='%1.0f%%', pctdistance=0.7, colors=["#FFFF33","#ff9100","#eaaa00","#6d6875"])
ax.add_artist(plt.Circle((0,0),0.3,fc='white'))
plt.show()
```

Del gráfico se observa que la mayor representación está dada por suburbios. Luego de eso, vienen los entornos rurales. Esto viene de la mano con la idea de hallar tendencias para aquellos estudiantes con menor acceso a los recursos educativos, como los podrían tener aquellos con mejor situación económica.

```python
districts.head(1)
```

```python
figure, axes = plt.subplots(ncols = 3, figsize = (18,5))
for idx, col in enumerate(['pct_black/hispanic', 'pct_free/reduced', 'county_connections_ratio']):
  index = districts[col].value_counts().index
  graph = sns.countplot(
      data= districts,
      x = col,
      order = index,
      palette = 'mako',
      ax = axes[idx% 3])

  total = len(districts[col])

  for p in graph.patches:
      percentage = '{:.1f}%'.format(100 * p.get_height()/total)
      x = p.get_x() + p.get_width() / 2 - 0.1
      y = p.get_height()
      graph.annotate(percentage, (x, y), ha='center', va='bottom')

  axes[idx%3].set_title(f'countplot for {col}')
plt.suptitle('countplots for districts')
plt.show()
```

De aquí se desprende:
- que la mitad de los distritos posee una identificación con negro/hispano de entre un 0 a 20%, mientras que solo un 3.4% de los distritos posee una identificación de entre 80% a 100%.
- Que en alrededor del 41% de los distritos, entre un 0 a 40% de los estudiantes son candidatos a una beca (parcial o total), mientras que un 23.2% los candidatos van del 40 al 100%. El resto corresponde a datos faltantes.
- Que por distrito, alrededor de un 69% de los PENDIENTE INTERPRETAR MEJOR ESTO

```python
index = districts['pp_total_raw'].value_counts().index
ax = sns.countplot(
      data= districts,
      y = 'pp_total_raw',
      order = index,
      palette = 'mako'
      )


total = len(districts['pp_total_raw'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width()/total)
    x = p.get_x() + p.get_width() + 0.1
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y), ha='left', va='center')

plt.title('countplots for per pupil total expenditure')
plt.show()
```
