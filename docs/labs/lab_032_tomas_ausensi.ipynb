3. Desarrolle una función `resumen_df(df)` para encontrar el total de elementos distintos y vacíos por columnas.

```python
# respuesta

def resumen_df(df):

    elementos_dis = [] #la idea es reunir los elementos únicos por cada columna
    elementos_vas = [] #La idea es reunir los elementos vacíos por cada columna

    nombres = df.columns

    for column in nombres: #recorre cada columna, recolectando elementos únicos y vacíos
      elementos_dis.append(df[column].nunique())
      elementos_vas.append(df[column].isnull().sum())

    result = pd.DataFrame({'nombres': nombres})
    result['elementos_distintos'] = elementos_dis
    result['elementos_vacios'] = elementos_vas

    return result
```

```python
# retornar
resumen_df(df)
```
